## Let's convert the data type of the "rate" coloumn to float and remove the denominator. 

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# Summary of the Data frame

```python
dataframe.info()
```

# Conclusion - There is no null value in the data frame.

# Types of Resturant: - 

```python
sns.countplot(x = dataframe['listed_in(type)'])
plt.xlabel("Types of resturant")
```

Conclusion: The majority of the resturants fall into the dining category.

## Dinning resturant are preferred by a large number of individulas.

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c = "green", marker = "o")
plt.xlabel("Type of resturant", c = "red", size = 20)
plt.ylabel("votes", c = "red", size = 20)
```

## The majority of resturants recived ratings

```python
plt.hist(dataframe['rate'], bins = 5)
plt.title("Ratings Distribution")
plt.show()
```

Conclusion - The maojority of resturants recieved ratings ranging from 3.5 to 4.

## The majority of couples prefer resturants with an approximate cost of 300 rupees.

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

## Wheather online orders recieved higher ratings than offline orders.

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)  
```

Conclusion - offline orders recieved lower  ratings in comparision to online orders, which obtained excellent ratings.

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()          
```

Conclusion - Dinning resturant primarily accept offline orders, whereas cafes primarily recive online orders. The sugests that client prefer to place order in person at resturants, but prefer online  ordering at cafes.
